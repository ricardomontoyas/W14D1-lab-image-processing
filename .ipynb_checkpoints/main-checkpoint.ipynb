### **Exercise: Image Creation and Color Manipulation**

*   Create a 100 x 100 image for each of the below visualization



*   Visualize the created images in a 1 x 3 subplot using matplotlib.


```python
import numpy as np
import matplotlib.pyplot as plt

# --- Creacin de las imgenes ---
# 1. Mitad vertical blanca y mitad vertical negra
img_vertical_split = np.zeros((100, 100), dtype=np.uint8)
img_vertical_split[:, :50] = 255  # mitad izquierda blanca
img_vertical_split[:, 50:] = 0    # mitad derecha negra

# 2. Mitad horizontal blanca y mitad horizontal negra
img_horizontal_split = np.zeros((100, 100), dtype=np.uint8)
img_horizontal_split[:50, :] = 0  # mitad superior blanca
img_horizontal_split[50:, :] = 255    # mitad inferior negra

# 3. Letra "L" negra de ancho 50 pxeles en y y x
img_L_shape = np.ones((100, 100), dtype=np.uint8) * 255  # fondo blanco
img_L_shape[:, 50:] = 0  # mitad izquierda blanca
img_L_shape[50:, :] = 0  # mitad izquierda blanca

# --- Visualizacin ---
fig, axes = plt.subplots(1, 3, figsize=(12, 4))
titles = ['Mitad Vertical', 'Mitad Horizontal', 'Forma L']
images = [img_vertical_split, img_horizontal_split, img_L_shape]

for ax, img, title in zip(axes, images, titles):
    ax.imshow(img, cmap='gray', vmin=0, vmax=255)
    ax.set_title(title)

    # Ticks de 0 a 100 cada 20
    ax.set_xticks(np.arange(0, 101, 20))
    ax.set_yticks(np.arange(0, 101, 20))

    # Cuadrcula interna
    ax.grid(color='red', linestyle='--', linewidth=0.5)

    # Lmites para ver todo el borde
    ax.set_xlim(0, 100)
    ax.set_ylim(100, 0)  # invertir eje Y

    # Borde exterior en negro
    ax.plot([0, 100, 100, 0, 0], [0, 0, 100, 100, 0], color='black', linewidth=2)

plt.tight_layout()
plt.show()



```

*   The color **red** looks too bright for the eyes. Isn't it?? Lets change the color and see how it appears.
    *    Create a new image where everything that is **'red' is changed to 'blue'**.
*   Visualize the original image and the created image using matplotlib subplot.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Step 1: Load image
img = cv2.imread(image_path)  # Or use the actual variable from your notebook
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Step 2: Reshape image to (num_pixels, 3)
pixels = img_rgb.reshape((-1, 3))

# Step 3: Apply KMeans
k = 5  # number of colors to cluster, adjust if needed
kmeans = KMeans(n_clusters=k, random_state=42, n_init='auto')
labels = kmeans.fit_predict(pixels)
centers = kmeans.cluster_centers_.astype(int)

# Step 4: Identify the 'red' cluster
# Red is roughly [255, 0, 0] in RGB. Let's calculate the distance to this target.
target_red = np.array([255, 0, 0])
distances_to_red = np.linalg.norm(centers - target_red, axis=1)
red_cluster_index = np.argmin(distances_to_red)

# Step 5: Replace red cluster with blue ([0, 0, 255])
new_centers = centers.copy()
new_centers[red_cluster_index] = [0, 0, 255]  # Blue in RGB

# Step 6: Recreate image with new colors
new_pixels = new_centers[labels]
new_img_rgb = new_pixels.reshape(img_rgb.shape).astype(np.uint8)

# Step 7: Show original and new image
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.title("Original Image")
plt.imshow(img_rgb)
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title("Red Replaced with Blue")
plt.imshow(new_img_rgb)
plt.axis('off')

plt.tight_layout()
plt.show()
```

### **Exercise: Image Operators**

*   You can find images 'model.png' and 'coat.png' in the images folder (First two images of the below visualization). Your task is to create an image from the given two images such a way that the model is wearing the coat (Third image in the visualization).
*   You can also find different textures in the images folder. Your task is to change the coat texture to any one of the given textures.
*   Visualize the images similar to the given visualization.

*Hint: Think masks!!!*



```python
import cv2
import numpy as np
import matplotlib.pyplot as plt


path = r"/Users/ricardomontoyasantos/Desktop/DSML/W14D1-lab-image-processing/images/"
coat = cv2.imread(path+'coat.png')         # Coat with black background
model = cv2.imread(path+'model.png')       # Model without coat
texture = cv2.imread(path+'texture2.png')  # Optional texture (like red coat pattern)

# Create mask from coat
gray_coat = cv2.cvtColor(coat, cv2.COLOR_BGR2GRAY)
_, mask = cv2.threshold(gray_coat, 10, 255, cv2.THRESH_BINARY)
mask_inv = cv2.bitwise_not(mask)

# Overlay coat onto model
model_bg = cv2.bitwise_and(model, model, mask=mask_inv)
coat_fg = cv2.bitwise_and(coat, coat, mask=mask)
model_with_coat = cv2.add(model_bg, coat_fg)

# Apply texture only on the coat area
textured_fg = cv2.bitwise_and(texture, texture, mask=mask)
model_with_textured_coat = cv2.add(model_bg, textured_fg)

# Convert to RGB for matplotlib
coat_rgb = cv2.cvtColor(coat, cv2.COLOR_BGR2RGB)
model_rgb = cv2.cvtColor(model, cv2.COLOR_BGR2RGB)
model_with_coat_rgb = cv2.cvtColor(model_with_coat, cv2.COLOR_BGR2RGB)
textured_model_rgb = cv2.cvtColor(model_with_textured_coat, cv2.COLOR_BGR2RGB)

#  Display results
plt.figure(figsize=(16, 6))

plt.subplot(1, 4, 1)
plt.title("Coat")
plt.imshow(coat_rgb)
plt.axis('off')

plt.subplot(1, 4, 2)
plt.title("Model")
plt.imshow(model_rgb)
plt.axis('off')

plt.subplot(1, 4, 3)
plt.title("Model with Coat")
plt.imshow(model_with_coat_rgb)
plt.axis('off')

plt.subplot(1, 4, 4)
plt.title("Model with Textured Coat")
plt.imshow(textured_model_rgb)
plt.axis('off')

plt.tight_layout()
plt.show()
```

# Contrast Enhancement

### **Exercise: Histogram Computation**

*   Read the **'astronaut' image** from data module.
*   Convert the image to grayscale.
*   Compute the **histogram of the image.** *Hint: histogram function is available in skimage.exposure package*
*   Plot the histogram using matplotlib plot.

```python

from skimage import data, color, exposure
import matplotlib.pyplot as plt

# Load the astronaut image
astronaut = data.astronaut()  # RGB image

#Convert to grayscale
gray_astronaut = color.rgb2gray(astronaut)

# Compute histogram
hist, bin_c = exposure.histogram(gray_astronaut)

# Plot histogram
plt.figure(figsize=(8, 6))
plt.bar(bin_c, hist, width=bin_c[1] - bin_c[0], edgecolor='red', align='center')
plt.title("Histogram of Grayscale Astronaut Image (256 Bins)")
plt.xlabel("Pixel Intensity")
plt.ylabel("Frequency")
plt.grid(True)
plt.tight_layout()
plt.show()
```

*   Change the bin count to 8 and compute the histogram of the image and plot the computed histogram using matplotlib plot.

```python
# solution

from skimage import data, color
import matplotlib.pyplot as plt
import numpy as np

# Step 3: Compute histogram with 8 bins
hist, bin_edges = np.histogram(gray_astronaut, bins=8, range=(0, 1))

# Step 4: Plot histogram
plt.figure(figsize=(8, 4))
plt.bar(bin_edges[:-1], hist, width=(bin_edges[1] - bin_edges[0]), edgecolor='black', align='edge')
plt.title("Histogram of Grayscale Astronaut Image (8 Bins)")
plt.xlabel("Pixel Intensity")
plt.ylabel("Frequency")
plt.grid(True)
plt.tight_layout()
plt.show()
```



*   What happens when you change the bin count? Does your inference change based on the bin count? If yes, then how do you define the correct bin count.




Changing the bin count in a histogram affects the level of detail shown in an images intensity distribution. A low bin count provides a simplified view by grouping many pixel values together, which may hide subtle variations, while a high bin count offers a more detailed and accurate representation. As a result, your interpretation can change depending on the bin count, especially in tasks like contrast enhancement or thresholding. The correct bin count depends on the goal: use fewer bins for quick summaries and more bins (like 256) when precision and detailed analysis are needed.

*   What happens when the bin count is very low and what happens when it is very high?

When the bin count is very low, the histogram becomes coarse and generalized, grouping a wide range of pixel intensities into few bins. This can hide important details, making it harder to detect subtle changes in brightness or contrast. In contrast, when the bin count is very high, the histogram becomes more detailed and can reveal fine variations in intensity, but may also appear noisy or cluttered, especially in images with limited pixel variation. The choice between low and high bin counts depends on whether a broad overview or detailed analysis is needed.
